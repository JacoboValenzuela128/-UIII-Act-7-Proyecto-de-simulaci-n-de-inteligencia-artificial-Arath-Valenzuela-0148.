# Maquillista y Modelo

```python
# ==========================================================
# SIMULADOR IA: BEAUTY MATCHER (Maquillista & Modelo)
# ==========================================================

print("==================================================")
print("     ¡BIENVENIDO A AI BEAUTY MATCHER v1.0!        ")
print("   El asistente inteligente para maquillistas    ")
print("==================================================")

# Variable para controlar que el menú se siga mostrando
opcion_menu = "0"

while opcion_menu != "3":
    # Impresión del Menú Principal
    print("\n--- MENÚ PRINCIPAL DE LA IA ---")
    print("1. Analizar Modelo y sugerir Look de Maquillaje")
    print("2. Calcular presupuesto de cosméticos para el Look")
    print("3. Salir de la app")

    opcion_menu = input("\nSelecciona una opción (1-3): ")

    # ------------------------------------------------------
    # OPCIÓN 1: ANALIZAR MODELO Y SUGERIR LOOK
    # ------------------------------------------------------
    if opcion_menu == "1":
        print("\n--- DIAGNÓSTICO DE LA MODELO ---")
        tono_piel = input("¿Qué tono de piel tiene la modelo? (clara / media / oscura): ")
        tipo_evento = input("¿El evento de la modelo es de (dia) o de (noche)?: ")

        print("\n[IA Analizando facciones y colores...]")

        # Lógica de la IA usando operadores lógicos (and) y condicionales
        if tono_piel == "clara" and tipo_evento == "dia":
            print("=> RECOMENDACIÓN IA: 'Look Fresh Peach'")
            print("   - Tonos: Sombras durazno, rubor rosado natural y labial gloss.")
            print("   - Tip: Usa base ligera o BB Cream para resaltar la piel.")

        elif tono_piel == "clara" and tipo_evento == "noche":
            print("=> RECOMENDACIÓN IA: 'Look Hollywood Glam'")
            print("   - Tonos: Labios rojos intensos y un delineado 'cat-eye' negro muy marcado.")
            print("   - Tip: Añade pestañas postizas de efecto 3D.")

        elif tono_piel == "media" and tipo_evento == "dia":
            print("=> RECOMENDACIÓN IA: 'Look Golden Hour'")
            print("   - Tonos: Sombras doradas y cafés cálidos con labios nude mate.")
            print("   - Tip: Un toque de iluminador dorado en los pómulos se verá increíble.")

        elif tono_piel == "media" and tipo_evento == "noche":
            print("=> RECOMENDACIÓN IA: 'Look Smokey Bronze'")
            print("   - Tonos: Ojos ahumados en tonos bronce y labios café delineados estilo 90s.")

        elif tono_piel == "oscura" and tipo_evento == "dia":
            print("=> RECOMENDACIÓN IA: 'Look Neon Pop'")
            print("   - Tonos: Un delineado de color brillante (azul o morado) y labios con mucho brillo.")
            print("   - Tip: Los rubores en tonos naranja o mandarina le darán mucha vida.")

        elif tono_piel == "oscura" and tipo_evento == "noche":
            print("=> RECOMENDACIÓN IA: 'Look Cyber Metallic'")
            print("   - Tonos: Sombras metálicas (plata o borgoña) y labios metálicos o gloss oscuro.")

        else:
            print("❌ Error: No escribiste las opciones exactamente como se pedían (clara/media/oscura o dia/noche).")

    # ------------------------------------------------------
    # OPCIÓN 2: CALCULAR PRESUPUESTO
    # ------------------------------------------------------
    elif opcion_menu == "2":
        print("\n--- CALCULADORA DE PRESUPUESTO PARA EL LOOK ---")
        cantidad_productos = input("¿Cuántos productos cosméticos vas a usar en la modelo?: ")
        cantidad_productos = int(cantidad_productos) # Convertimos a número entero

        precio_promedio = input("¿Cuál es el precio promedio de tus productos en dólares ($)?: ")
        precio_promedio = float(precio_promedio) # Convertimos a número con decimales

        # Cálculo básico
        total_estimado = cantidad_productos * precio_promedio
        print(f"\nCosto total estimado en cosméticos: ${total_estimado:.2f}")

        # Alerta de la IA según el costo
        if total_estimado > 100:
            print("⚠️ IA Alerta: Tu presupuesto es elevado. ¡Asegúrate de cobrar bien tu trabajo de maquillista!")
        else:
            print("✅ IA Info: Presupuesto accesible para prácticas o sesiones editoriales estándar.")

    # ------------------------------------------------------
    # OPCIÓN 3: SALIR
    # ------------------------------------------------------
    elif opcion_menu == "3":
        print("\nCerrando el sistema...")

    # CONDICIONAL POR SI ESCRIBEN OTRA COSA EN EL MENÚ
    else:
        print("❌ Opción no válida. Por favor, elige 1, 2 o 3.")

    print("\n--------------------------------------------------")

print("\n==================================================")
print(" ¡Gracias por usar la IA de Maquillaje! ¡Éxito en la pasarela! ")
print("==================================================")
```
